```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv('d.csv')
print(df.info())
```

```python
# diagramas de cajas y bigotes,

fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 8))

sns.boxplot(data=df, x='Agricultura.ganaderia', y='Periodo', ax=axes[0, 0])
sns.boxplot(data=df, x='Industria.electricidad', y='Periodo', ax=axes[0, 1])
sns.boxplot(data=df, x='Industria.manufacturera', y='Periodo', ax=axes[1, 0])
sns.boxplot(data=df, x='Construccion', y='Periodo', ax=axes[1, 1])

axes[0, 0].set_title('Agricultura.ganaderia')
axes[0, 1].set_title('Industria.electricidad')
axes[1, 0].set_title('Industria.manufacturera')
axes[1, 1].set_title('Construccion')

plt.tight_layout()

plt.show()
```

```python
# histogramas

fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 8))

# 'Agricultura.ganaderia'
axes[0, 0].hist(df['Agricultura.ganaderia'], bins=10, color='blue', alpha=0.7)
axes[0, 0].set_title('Agricultura.ganaderia')

# 'Industria.electricidad'
axes[0, 1].hist(df['Industria.electricidad'], bins=10, color='green', alpha=0.7)
axes[0, 1].set_title('Industria.electricidad')

# 'Industria.manufacturera'
axes[1, 0].hist(df['Industria.manufacturera'], bins=10, color='pink', alpha=0.7)
axes[1, 0].set_title('Industria.manufacturera')

# 'Construccion'
axes[1, 1].hist(df['Construccion'], bins=10, color='purple', alpha=0.7)
axes[1, 1].set_title('Construccion')


plt.tight_layout()
plt.show()
```

```python
# mapa de calor

corr = df.corr()
sns.heatmap(data = corr, vmin = -1, vmax = 1, cmap = 'YlOrRd',annot=True)
```
